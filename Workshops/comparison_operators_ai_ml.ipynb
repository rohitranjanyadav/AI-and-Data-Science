# Comparison Operators in Python for AI & ML
This notebook demonstrates the 6 comparison operators in Python, with context-specific examples tailored for Artificial Intelligence and Machine Learning applications (such as thresholding, accuracy evaluation, and data filtering).

## 1. Equal To (`==`)
Used to check if two values are identical. In ML, this is frequently used to evaluate classification performance by comparing predicted labels to ground-truth labels.

```python
# Example: Checking if a model's prediction matches the actual label
true_label = "Cat"
predicted_label = "Cat"

is_correct = (predicted_label == true_label)
print(f"Prediction correct? {is_correct}")
```

## 2. Not Equal To (`!=`)
Checks if two values are different. In AI pipelines, you can use this to filter out missing values or unlabelled data points during preprocessing.

```python
# Example: Filtering out corrupt/missing tokens in NLP preprocessing
token = "NaN"
is_valid_data = (token != "NaN")

print(f"Is the data point valid to use? {is_valid_data}")
```

## 3. Greater Than (`>`)
Checks if the left value is strictly larger than the right value. Commonly used to implement strict thresholds, like ensuring an optimization step yields an improvement greater than a minimum tolerance.

```python
# Example: Early stopping when validation loss stops dropping significantly
previous_loss = 0.54
current_loss = 0.55

loss_increased = (current_loss > previous_loss)
print(f"Should we consider early stopping? {loss_increased}")
```

## 4. Less Than (`<`)
Checks if the left value is strictly smaller than the right value. Often used to verify if error metrics or loss functions have fallen below an acceptable threshold.

```python
# Example: Verifying if Mean Squared Error (MSE) meets requirements
mse_error = 0.035
acceptable_threshold = 0.05

performance_met = (mse_error < acceptable_threshold)
print(f"Is the model error within acceptable limits? {performance_met}")
```

## 5. Greater Than or Equal To (`>=`)
Checks if the left value is larger than or equal to the right value. In Computer Vision, confidence scores from object detection models are compared against a threshold using this operator before drawing bounding boxes.

```python
# Example: Confidence filtering for Object Detection models
confidence_score = 0.85
min_confidence_required = 0.85

display_bounding_box = (confidence_score >= min_confidence_required)
print(f"Display bounding box to user? {display_bounding_box}")
```

## 6. Less Than or Equal To (`<=`)
Checks if the left value is smaller than or equal to the right value. Used frequently in resource-constrained environments to make sure model file sizes or inference latency match hardware constraints.

```python
# Example: Checking if model size fits edge computing memory capacity
model_size_mb = 45.2
max_allowed_mb = 50.0

fits_on_device = (model_size_mb <= max_allowed_mb)
print(f"Can this model deploy to the edge device? {fits_on_device}")
```
